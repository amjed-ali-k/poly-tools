```python
#  ----- Author Details  ---------
#          Amjed Ali K
#  https://github.com/amjed-ali-k
#  -------------------------------


# DONT TOUCH CODES BELOW
import csv
import json
import os
import click


# SECTION 0

# List existing csv files in current directory and choose one of them as input file
# Ask input of Semester (1 - 6), Exam Month (Nov, Apr), Year
# Ask if Supplymentry is to be excluded
# Add validation for inputs



# List existing csv files in current directory and select one of them as input file using arrow key
input_file = click.prompt('Select Input File', type=click.Choice(os.listdir('.'), case_sensitive=False), show_choices=False)

# Ask input of Semester (1 - 6) and validate using click
semester = click.prompt('Enter Semester (1 - 6)', type=click.IntRange(1, 6), show_choices=False)

# Ask input of Exam Month (Nov, Apr) and validate using click
exam_month = click.prompt('Enter Exam Month (Nov, Apr)', type=click.Choice(['Nov', 'Apr'], case_sensitive=False), show_choices=False)

# Ask input of Year (2021, 2022, 2023, 2024) and validate using click
year = click.prompt('Enter Year (2021, 2022, 2023, 2024)', type=click.Choice(['2021', '2022', '2023', '2024', '2025', '2026', '2027', '2028', '2029', '2030'], case_sensitive=False), show_choices=False)

# Ask if Supplymentry is to be excluded and validate using click
exclude_supplymentry = click.prompt('Exclude Supplymentry (Y/N)', type=click.Choice(['Y', 'N'], case_sensitive=False), show_choices=False)

# Convert exclude_supplymentry to boolean
if exclude_supplymentry == 'Y':
    exclude_supplymentry = True
else:
    exclude_supplymentry = False

# Create output file name
output_file = 'Sorted_S' + semester + '_Result-' + exam_month + '-' + year + '-Grades.csv'
grades_sheet = 'Analysis-S' + semester + '-' + exam_month + '-' + year + '.csv'






# SECTION I

data = {}
f = open(input_file, 'r')
reader = csv.reader(f)
next(reader)
subjects = []
for item in reader:
    if exclude_supplymentry and item[5] != 'Regular':
        continue
    if not item[0] in data:
        data[item[0]] = {
            "Branch": item[2],
            "Name": item[1],
            "Semester": item[3],
        }
    if item[9] == "F" and item[6] == 'Absent':
        data[item[0]][item[4]] = 'Absent'
    else:
        data[item[0]][item[4]] = item[9]
    if item[7] == "Withheld":
        data[item[0]][item[4]] = "Withheld"
       
    if item[7] == "With held for Malpractice":
        data[item[0]][item[4]] = "F"
   
    if item[4] not in subjects:
        subjects.append(item[4])
f.close()

#  SECTION II

f = open(output_file, 'w')
writer = csv.writer(f)
writer.writerow(['RegNo', 'Name', 'Branch', 'Semester'
] + subjects)
for key in data:
    row = []
    row.append(key)
    row.append(data[key]['Name'])
    row.append(data[key]['Branch'])  
    row.append(data[key]['Semester'])
    for sub in subjects:
        if sub in data[key]:
            row.append(data[key][sub])
        else:
            row.append('Not Found')
    writer.writerow(row)
   
   
# SECTION III


status = {}
for sub in subjects:
    status[sub] = {}
for key in data:
    val = data[key]
    for sub in subjects:
        if sub in val:
       
            if val[sub] in status[sub]:
                status[sub][val[sub]] += 1
            else:
                status[sub][val[sub]] = 1

# SECTION IV

f = open(grades_sheet, 'w')
writer = csv.writer(f)
grades = ['S', 'A', 'B', 'C', 'D', 'E', 'F', 'Absent', 'Withheld']
print("FILES CREATED SUCCESSFULLY")
print("==> GRADE STATUS <==")

writer.writerow(['Subject',
] + grades)
for key in status:
    row = []
    row.append(key)
    for grade in grades:
        if grade in status[key]:
            row.append(status[key][grade])
        else:
            row.append(0)
    print(row)
    writer.writerow(row)
   
   
total_passed = 0
total_failed = 0
total_withheld = 0
for key in data:
    isFailed = False
    isWithheld = False
    for sub in data[key]:
        if sub in subjects:
            if data[key][sub] == 'F' or data[key][sub] == 'Absent':
                isFailed = True
            if data[key][sub] == 'Withheld':
                isWithheld = True
    if isFailed:
        total_failed += 1
    elif isWithheld:
        total_withheld += 1
    else:
        total_passed += 1

writer.writerow(["Total Passed", total_passed])
writer.writerow(["Total Failed", total_failed])
writer.writerow(["Total Withheld", total_withheld])
writer.writerow(["Pass Percentage", (total_passed /(total_failed + total_passed)) * 100 ])


f.close()         
```
